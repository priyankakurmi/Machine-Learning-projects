```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
def gradient(data,m,alpha):
    m_slope=np.zeros(data.shape[1]-1)
    M=len(data)
    for i in range (M):
        for j in range (data.shape[1]-1): 
          x=data[i,0:data.shape[1]-1]
          y=data[i,data.shape[1]-1]
          m_slope[j]+=(-2/M)*((y-(m*x).sum())*data[i][j])
          m[j]=m[j]-(alpha*m_slope[j])
    return m
```

```python
def cost(m,data):
    cost=0
    M=len(data)
    for i in range (M):
           x=data[i,0:data.shape[1]-1]
           y=data[i,data.shape[1]-1]
           cost+=(1/M)*((y-(m*x).sum())**2)
    return cost
```

```python
def iteration(data,number_iteration):
    m=np.zeros(data.shape[1]-1)
    alpha=0.0001
    for i in range (number_iteration):
        m=gradient(data,m,alpha)
        print('total_cost:',i, cost(m,data))
    return m
```

```python
def run():
    testing=pd.read_csv('train.csv',delimiter=',',skipinitialspace=True)
    data2=pd.DataFrame(testing)
    y_train=data2["EP"]
    data2.drop("EP",inplace=True,axis=1)
    data2['T_square']=data2['# T']**2
    data2['V_square']=data2['V']**2
    scaler.fit(data2)
    data1=scaler.transform(data2)
    data=np.array(data1)
    #arr=np.ones(data.shape[0]).reshape(-1,1)
    data=np.append(data,np.ones(data.shape[0]).reshape(-1,1),axis=1)
    data=np.append(data,np.array(y_train).reshape(-1,1),axis=1)
    number_iteration=350
    m=iteration(data,number_iteration)
    return m
```

```python
m1=run()
```

```python
#m=m1.reshape(m1.shape[0],)
#dele=m.shape[0]-2
#m=np.delete(m,dele)
test=pd.read_csv('test.csv',delimiter=',',skipinitialspace=True,header=None)
test_data=np.array(test)
test_data=np.append(test_data,(test_data[:,0]**2).reshape(-1,1),axis=1)
test_data=np.append(test_data,(test_data[:,1]**2).reshape(-1,1),axis=1)
test_data_scale=scaler.transform(test_data)
test_data_scale=np.append(test_data_scale,np.ones(test_data_scale.shape[0]).reshape(-1,1),axis=1)
y_predict=np.zeros(len(test_data_scale))
for k in range (len(test_data_scale)):
    y_predict[k]=(test_data_scale[k]*m1).sum()
np.savetxt('Boston Gradient Descent prediction.csv',y_predict,fmt='%.5f',delimiter=',')
```
